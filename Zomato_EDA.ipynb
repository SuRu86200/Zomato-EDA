```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['rate', 'location', 'rest_type', 'cuisines', 'approx_cost(for two people)'], inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df['rate'] = df['rate'].str.replace('/5','').str.strip()
```

```python
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
print(df['approx_cost(for two people)'].unique(),[20])
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','').str.strip()
```

```python
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')
```

```python
df['phone'] = df['phone'].str.split('\r\n').str[0].str.strip()
```

```python
df.rename(columns={'approx_cost(for two people)' : 'cost', 'listed_in(type)' : 'type', 'listed_in(city)' : 'city'}, inplace=True)
```

```python
df.dtypes
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
print("Clean data shape:", df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Top 10 Locations by Restaurants
plt.figure(figsize=(12,5))
df['location'].value_counts().head(10).plot(kind='bar', color='steelblue')
plt.title('top 10 locations by numbers of restraunts')
plt.xlabel('location')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart1_locations.png', bbox_inches='tight')
plt.show()
print("observation : Btm has highest number of restraunt in Banglore")
```

```python
# Online Order vs Rating
plt.figure(figsize=(7,12))
sns.boxplot(x='online_order', y='rate', data=df)
plt.title('online_order vs rating')
plt.xlabel('online order available')
plt.ylabel('rating')
plt.tight_layout()
plt.savefig('chart2_online_orders vs rating.png', bbox_inches='tight')
plt.show()
print('Observation: Restaurants with online ordering have higher ratings')
```

```python
#Top 10 Cuisines
plt.figure(figsize=(12,5))
df['cuisines'].value_counts().head(10).plot(kind='bar', color='coral')
plt.title('top 10 most popular cuisines')
plt.xlabel('cuisines')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart3_Top 10 cuisines.png', bbox_inches='tight')
plt.show()
print("observation : north indian food is very popular in bangalore")
```

```python
#Restaurant Type Distribution
plt.figure(figsize=(8,5))
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('restraunts type distribution')
plt.ylabel('')
plt.tight_layout()
plt.savefig('chart4_rest types.png', bbox_inches='tight')
plt.show()
```

```python
#Cost Distribution
plt.figure(figsize=(8,5))
df['cost'].plot(kind='hist', bins=30, color='pink', edgecolor='black')
plt.title('distribution of cost for two people')
plt.xlabel('cost (INR)')
plt.ylabel('number of restraunts')
plt.tight_layout()
plt.savefig('chart5_avrg cost.png', bbox_inches='tight')
plt.show()
print("Observation: Most restaurants cost between 200-600 INR for two")
```

```python
#Top 10 Locations by Average Rating
plt.figure(figsize=(12,5))
df.groupby('location')['rate'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='purple')
plt.title('top 10 locations by average rating')
plt.xlabel('location')
plt.ylabel('avg_rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart6_avg rating locations.png', bbox_inches='tight')
plt.show()
print("Observation: Lavelle Road has highest average restaurant rating")
```

```python
df.to_csv('zomato_cleaned.csv', index=False)
print ("saved as csv")
```
